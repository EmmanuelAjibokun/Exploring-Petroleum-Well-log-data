```python
fig, axes = plt.subplots(figsize=(10,20))

# list of log labels
curve_names = ["Gamma (gAPI)", "Density (g/cm3)", "Neutron (m3/m3)", "Resistivity (ohm.m)"]
curve_colors = ["green", "red", "blue", "black"]

# Set up the plot axes
ax1 = plt.subplot2grid((1,3), (0,0), rowspan=1, colspan=1)
ax2 = plt.subplot2grid((1,3), (0,1), rowspan=1, colspan=1)
ax3 = ax2.twiny()
ax4 = plt.subplot2grid((1,3), (0,2), rowspan=1, colspan=1)

# Set up individual subplots
ax1.plot("GR", "DEPT", data=las_df, color="green", lw=0.5)
ax1.set_xlim(0, 500)

ax2.plot("RHOB", "DEPT", data=las_df, color="red", lw=0.5)
ax2.set_xlim(1.3, 3.0)

ax3.plot("NPHI", "DEPT", data=las_df, color="blue", lw=0.5)
ax3.set_xlim(0.9, 0.02)

ax4.plot("RDEP", "DEPT", data=las_df, color="black", lw=0.5)
ax4.set_xlim(0.2, 200)
ax4.semilogx()

# Setting up common elements between each subplots
for i, ax in enumerate(fig.axes):
    ax.set_ylim(3300, 25)
    ax.set_xlabel(curve_names[i], color=curve_colors[i])
    ax.xaxis.set_ticks_position("top")
    ax.xaxis.set_label_position("top")
    if i == 2:
        ax.spines["top"].set_position(("axes", 1.04))
        continue
    ax.grid()

for ax in [ax1, ax4]:
    plt.setp(ax.get_yticklabels(), visible=False)

plt.savefig("well-1-log-plot.jpg", dpi=200)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,20))

# set up the plot axis
ax[0].plot(las_df["SP"], las_df["DEPT"], color="green")
ax[1].plot(las_df["DTC"], las_df["DEPT"], color="red")

# Setting up the spont. potential log
ax[0].set_xlabel("Sponteneous Potential")
ax[0].set_xlim(-150, 154)
ax[0].set_ylim(3300, 25)
ax[0].grid()

# Setting up the shear delta time
ax[1].set_xlabel("Shear Delta Time")
ax[1].set_xlim(56, 207)
ax[1].set_ylim(3300, 25)
ax[1].grid()
```
