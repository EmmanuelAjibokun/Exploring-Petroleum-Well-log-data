```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import lasio
```

```python
df = las.df()
df.describe()
```

```python
# with the depth being our initial index, a new index column is created, pushing dept column as the first curve.
df.reset_index(inplace=True)
```

```python
df.head()
```

```python
df.info()
```

#### Create X (features matrix)

```python
# dropping the target curve
X = df.drop("GR", axis=1)
```

```python
# drop curves with too little entries
X = X.drop(["RSHA", "RXO"], axis=1)
```

```python
X.info()
```

#### Create y (target)

```python
y = df["GR"]
```

```python
y.info()
```

### Option1: Fill missing data with Pandas

```python
# Fill missing data in features curves with mean of respective curve
for curve in X:
    if curve != "DEPT":
        X[curve].fillna(X[curve].mean(), inplace=True)
```

```python
X.isna().sum()
```

```python
y.fillna(y.mean(), inplace=True)
y.isna().sum()
```

### Build machine learning model
    Following our scikit-learn machine learning map, we have been able to narrow our model to a few Regression model:
   #### RidgeRegression

```python
# Import algorithm/estimator
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

# set up random seed
np.random.seed(42)

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Instantiate and fit the model (on the train set)
reg = Ridge()

# fitting
reg.fit(X_train, y_train)

# Score model
reg.score(X_test, y_test)
```

```python
# train model with 5 different part of the same data
cross_val_score(reg, X, y)
```

#### SVR(Support Vector Regression) model

```python
from sklearn.svm import SVR
regr = SVR()
regr.fit(X_train, y_train)
regr.score(X_test, y_test)
```

```python
cross_val_score(regr, X, y)
```

#### Linear regression

```python
from sklearn.linear_model import LinearRegression
l_reg = LinearRegression()
l_reg.fit(X_train, y_train)
l_reg.score(X_test, y_test)
```

```python
cross_val_score(l_reg, X, y)
```
