### Dynamically add the units of the columns

```python
curves = las.curves
for curve_item in curves:
    for column in df:
        if curve_item.mnemonic == column:
            df.rename(
                columns={column: f"{column} ({curve_item.unit})"},
                inplace=True)
```

```python
df
```

We can see from the returned results that we have several columns of data, each column represents measurements that have been taken whilst the logging tools have been moved along the well.

The some of the columns represent the following:

- CALI for borehole caliper
- RHOB for bulk density
- GR for gamma ray
- NPHI for neutron porosity
- RDEP for deep resisitivity
- RMED for medium resistivity
- SP for sponteneous potential
- DTC for compress. delta time

To make it easier to work with our dataframe, we have converted the dataframe index, which is set to depth, to a column within the dataframe. We achieved this by reseting the index like so.

Note that inplace=True allows us to make the changes to the original dataframe object.

```python
fig, axes = plt.subplots(figsize=(10,20))

# list of log labels
curve_names = ["Gamma (gAPI)", "Density (g/cm3)", "Neutron (m3/m3)", "Resistivity (ohm.m)"]
curve_colors = ["green", "red", "blue", "black"]

# Set up the plot axes
ax1 = plt.subplot2grid((1,3), (0,0), rowspan=1, colspan=1)
ax2 = plt.subplot2grid((1,3), (0,1), rowspan=1, colspan=1)
ax3 = ax2.twiny()
ax4 = plt.subplot2grid((1,3), (0,2), rowspan=1, colspan=1)

# Set up individual subplots
ax1.plot("GR (gAPI)", "DEPT (m)", data=df, color="green", lw=0.5)
ax1.set_xlim(7.3, 154)

ax2.plot("RHOB (g/cm3)", "DEPT (m)", data=df, color="red", lw=0.5)
ax2.set_xlim(1.6, 2.7)

ax3.plot("NPHI (m3/m3)", "DEPT (m)", data=df, color="blue", lw=0.5)
ax3.set_xlim(0.6, 0.2)

ax4.plot("RDEP (ohm.m)", "DEPT (m)", data=df, color="black", lw=0.5)
ax4.set_xlim(0.2, 20.6)
ax4.semilogx()

# Setting up common elements between each subplots
for i, ax in enumerate(fig.axes):
    ax.set_ylim(2314, 96)
    ax.set_xlabel(curve_names[i], color=curve_colors[i])
    ax.xaxis.set_ticks_position("top")
    ax.xaxis.set_label_position("top")
    if i == 2:
        ax.spines["top"].set_position(("axes", 1.04))
        continue
    ax.grid()

for ax in [ax1, ax4]:
    plt.setp(ax.get_yticklabels(), visible=False)
```

## Creating Histograms using matplotlib

```python
plt.hist(df['GR (gAPI)'])
plt.show()
```

```python
mean = df['GR (gAPI)'].mean()
p5 = df['GR (gAPI)'].quantile(0.05)
p95 = df['GR (gAPI)'].quantile(0.95)

plt.hist(df['GR (gAPI)'], bins=30,color='red', alpha=0.5, edgecolor="black")
plt.xlabel('Gamma Ray - API', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xlim(0, 151)

plt.axvline(mean, color='blue', label='mean')
plt.axvline(p5, color='green', label='5th Percentile')
plt.axvline(p95, color='yellow', label='95th Percentile')

plt.legend()
plt.show()
```
